```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from google.colab import files
```

```python
# Upload file
uploaded = files.upload()

# Load the uploaded CSV
df = pd.read_csv(list(uploaded.keys())[0])
df.head()
```

```python
print(df.info())
print(df.head())
```

```python
# Group items by Member_number and create a list of items per transaction
transactions = df.groupby('Member_number')['itemDescription'].apply(list).tolist()

# Check first 5 transactions
transactions[:5]
```

```python
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
basket_df = pd.DataFrame(te_array, columns=te.columns_)

# Preview the encoded dataframe
basket_df.head()
```

```python
frequent_itemsets = apriori(basket_df, min_support=0.05, use_colnames=True)
frequent_itemsets.sort_values(by='support', ascending=False, inplace=True)

# Show top 10 frequent itemsets
frequent_itemsets.head(10)
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)
rules = rules.sort_values(by='lift', ascending=False)

# Show top 10 rules
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(10)
```

```python
for _, row in rules.iterrows():
    print(f"Bundle: {list(row['antecedents'])} -> {list(row['consequents'])} | Confidence: {row['confidence']:.2f} | Lift: {row['lift']:.2f}")
```
